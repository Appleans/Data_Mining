```python
import pandas as pd

# Load the .ods file
df = pd.read_excel(r'C:\D\TUGAS-TUGAS ANGGUN (SK)\Semester 3\Data mining\Dataset DDOS IPv6 2022.ods', engine='odf')

# Print the DataFrame
print(df)
```

```python
# Menampilkan 5 baris pertama
print("5 Baris Pertama:")
print(df.head())

# Menampilkan 5 baris terakhir
print("\n5 Baris Terakhir:")
print(df.tail())
```

```python
# Cek nilai missing (hilang)
missing_values = df.isnull().sum()
print("Nilai Missing di setiap kolom:")
print(missing_values)
```

```python
import numpy as np

# Menghitung Z-score
from scipy import stats

z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
outliers = (z_scores > 3).sum(axis=0)
print("Jumlah outliers di setiap kolom numerik:")
print(outliers)
```

```python
# Menampilkan kolom numerik
numerical_columns = df.select_dtypes(include=[float, int]).columns
print("Kolom Numerik dalam Dataset:")
print(numerical_columns)
```

```python
# Daftar kolom yang ingin dianalisis
kolom = ['Dur','sMeanPktSz', 'dMeanPktSz', 'RunTime', 'SrcRate', 'DstRate']

# Memilih hanya kolom numerik dari DataFrame yang diinginkan
numerik_df = df[kolom].select_dtypes(include=[float, int])

# Menghitung mean untuk kolom yang dipilih
mean = numerik_df.mean()
print("Mean:")
print(mean)

# Menghitung median untuk kolom yang dipilih
median = numerik_df.median()
print("\nMedian:")
print(median)

# Menghitung mode untuk kolom yang dipilih
# Mengambil mode pertama jika ada beberapa mode
mode = numerik_df.mode().iloc[0]
print("\nModus:")
print(mode)

# Menghitung standar deviasi untuk kolom yang dipilih
std_dev = numerik_df.std()
print("\nStandar Deviasi:")
print(std_dev)

# Menghitung variansi untuk kolom yang dipilih
variance = numerik_df.var()
print("\nVariansi:")
print(variance)

# Menghitung skewness (kemiringan distribusi) untuk kolom yang dipilih
skewness = numerik_df.skew()
print("\nSkewness:")
print(skewness)

# Menghitung kurtosis (keruncingan distribusi) untuk kolom yang dipilih
kurtosis = numerik_df.kurtosis()
print("\nKurtosis:")
print(kurtosis)
```
